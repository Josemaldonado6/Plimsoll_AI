## 2. DATASET CONFIGURATION
Upload your dataset (images + labels) to Google Drive or use a sample.

```python
# Mount Drive
from google.colab import drive
drive.mount('/content/drive')

# CONFIG: Path to data.yaml
data_path = '/content/drive/MyDrive/Plimsoll_Data/data.yaml'
```

## 3. TRAINING (The Learning Phase)
We use YOLOv11n (Nano) for max speed on Edge or YOLOv11s (Small) for balance.

```python
from ultralytics import YOLO

# Load Pre-trained Model
model = YOLO('yolo11n.pt') 

# Train
# imgsz=640 is standard. epochs=100 for decent convergence.
results = model.train(data=data_path, epochs=100, imgsz=640, plots=True)
```

## 4. EXPORT TO INTEL OPENVINO (The Optimization)
**CRITICAL:** This step converts the model to run on your ASUS Zenbook NPU.

```python
# Export to OpenVINO format (FP16 for NPU Acceleration)
path = model.export(format='openvino', half=True)
print(f"Model exported to: {path}")
```
